Solvers:

```r
# Normalised amplitude solver
solveX <- function(freq, data, A, B) {
    output <- data.frame(matrix(NA, nrow = nrow(A), ncol = 0))
    for (f in freq) {
        output[as.character(f)] <- solve(apply(X = A, MARGIN = c(1,2), FUN = function(x) {eval(parse(text = x))}),
                                         apply(X = B, MARGIN = c(1), FUN = function(x) {eval(parse(text = x))}))
    }
    return(output)
}
```

Input impedance may be obtained from the equation:

$$
\hat Z_{in}
=
\chi_g
\left\{
    1
    +
    j
    Q_1
    \left[
        \frac {f}{f_1}
        -
        \frac{f_1}{f}
        -
        \left(
            \frac{f}{f_2}
            \frac{K_{12}^H}{2}
            -
            \frac{f_1}{f}
            \frac{K_{12}^E}{2}
        \right)
        \frac{\dot X_2}{\dot X_1}
     \right]
\right\}^{-1}
$$

where $\dot X_n$ are the normalized complex amplitudes in cells

```r
z_fun <- "data$chi_g1/(1+1i*data$Q1*(f/data$f1-data$f1/f-(f*data$Kh12/(data$f2*2)-data$f1*data$Ke12/(f*2))*X2_norm))"

# Input impedance solver
solveZ <- function(X) {
    output <- list(f=NULL, z=NULL)
    freq <- as.numeric(names(X))
    X <- apply(X=X, MARGIN = c(2), FUN = function(x) {return(x/x[1])})
    for (f in freq) {
        X2_norm <- X[,as.character(f)][2]
        output$f <- append(f, output$f)
        output$z <- append(eval(parse(text = z_fun)), output$z)
    }
    return(output)
}
```

## Comparison without transmission line

To test equations, lets solve the problem without transmission line.

For this we will take first three equations from the system above.
In place of the transmission line the terminating load is placed.

```r
A <- matrix(c(a11, a21, a31,
              a12, a22, a32,
              a13, a23, a33), nrow = 3)
B <- matrix(c(b1, b2, b3), nrow = 3)

cat("A:\n")
print(A)
cat("\n\nB:\n")
print(B)
```

Input impedance of such a device is shown below.

```r
Z <- solveZ(solveX(seq(2850e6, 2860e6, length.out = 201), data, A, B))
Z_db <- 20 * log10(abs(Mod((Z$z - 1) / (Z$z + 1))))  # S11_lin = (Z-1)/(Z+1); S11_dB = 20*log10(S11_lin)
plot(x=Z$f/1e6, y=Z_db, type = 'l', ylab = "S11(dB)", xlab = "f, MHz")
grid(col='black')
```

## Comparison with transmission line

Now we will take two devices and connect them via waveguide.

|![fem model](./static/model_wg.png)|

|Model two devices connected via waveguide in FEM software|

```r
A <- matrix(c(a11, a21, a31, a41, a51, a61,
              a12, a22, a32, a42, a52, a62,
              a13, a23, a33, a43, a53, a63,
              a14, a24, a34, a44, a54, a64,
              a15, a25, a35, a45, a55, a65,
              a16, a26, a36, a46, a56, a66), nrow = 6)
B <- matrix(c(b1, b2, b3, b4, b5, b6), nrow = 6)


cat("A:\n")
print(A)
cat("\n\nB:\n")
print(B)
```

```r
dat <- seq(0.04,0.24,0.01)
data$L <- dat[1]
Z <- solveZ(solveX(seq(2850e6, 2860e6, length.out = 201), data, A, B))
Z_db <- 20 * log10(abs(Mod((Z$z - 1) / (Z$z + 1))))  # S11_lin = (Z-1)/(Z+1); S11_dB = 20*log10(S11_lin)
plot(x=Z$f/1e6, y=Z_db, type = 'l', ylab = "S11(dB)", xlab = "f, MHz", ylim = c(-60, 0))
j <- 1
for (i in dat[-1]) {
data$L <- i
j <- j+1
Z <- solveZ(solveX(seq(2850e6, 2860e6, length.out = 201), data, A, B))
Z_db <- 20 * log10(abs(Mod((Z$z - 1) / (Z$z + 1))))  # S11_lin = (Z-1)/(Z+1); S11_dB = 20*log10(S11_lin)
lines(x=Z$f/1e6, y=Z_db, type = 'l', col=j)
}
grid(col='black')
#legend("bottomright", legend=round(dat,2), lty=1, col=c(1:length(dat)))
```
